```python
from matplotlib import pyplot as plt
import numpy as np
```

$f(x) = \frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}} $

```python
x = np.linspace(-10, 10, 1000)
y = 1/np.sqrt(2*np.pi)*np.exp(-x**2/2)
plt.plot(x, y, color='Red', label=r'$f(x) = \frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}} $')
plt.xlabel('X o\'q')
plt.ylabel('Y o\'q')
plt.title('Grafik')
plt.legend()
```
